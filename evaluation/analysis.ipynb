```python
import pandas as pd
import json
import matplotlib.pyplot as plt
from collections import Counter
results = pd.read_csv('evaluation/results.csv')
with open('evaluation/summary.json','r') as f:
    summary = json.load(f)
summary
```

```python
intent_cm = pd.crosstab(results['true_intent'], results['pred_intent'])
plt.figure(figsize=(8,6))
plt.imshow(intent_cm, cmap='Blues')
plt.title('Intent Confusion Matrix')
plt.xticks(range(len(intent_cm.columns)), intent_cm.columns, rotation=45, ha='right')
plt.yticks(range(len(intent_cm.index)), intent_cm.index)
plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
results['processing_time_ms'].plot(kind='hist', bins=20)
plt.title('Processing Time Distribution')
plt.xlabel('ms')
plt.show()
```

```python
sent_cm = pd.crosstab(results['true_sentiment'], results['pred_sentiment'])
plt.figure(figsize=(8,6))
plt.imshow(sent_cm, cmap='Reds')
plt.title('Sentiment Confusion Matrix')
plt.xticks(range(len(sent_cm.columns)), sent_cm.columns, rotation=45, ha='right')
plt.yticks(range(len(sent_cm.index)), sent_cm.index)
plt.colorbar()
plt.tight_layout()
plt.show()
```
