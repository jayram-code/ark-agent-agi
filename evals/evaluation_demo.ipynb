This notebook runs the evaluation harness and visualizes the performance of the agent system.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import sys
import os

# Run the harness (simulated call)
!python evals/harness.py

# Load results
with open('evals/results.json', 'r') as f:
    data = json.load(f)

df = pd.DataFrame(data)
df.head()
```

```python
# Visualize Pass Rate
pass_rate = df['passed'].mean() * 100
print(f"Pass Rate: {pass_rate:.1f}%")

df['passed'].value_counts().plot(kind='bar', color=['green', 'red'])
plt.title('Evaluation Results')
plt.xlabel('Passed')
plt.ylabel('Count')
plt.show()
```
